**6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.**

```python
def get_median(sample):
 n=len(sample)     
 index= n//2
 if n%2:        # sample with even number of obervation
   return sorted[index]

 return sum(sorted(sample)[index-1:index+1])/2  #sample with odd number of obervation
```

```python
get_median(housing['longitude']) # getting median using user defined function
```

**10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**



```python
#condition
case=[(housing['total_bedrooms']<=10),
      (housing['total_bedrooms']>=11)&(housing['total_bedrooms']<1000),
       (housing['total_bedrooms']>1000)]

#choicelist
size=['small','medium','large']

#numpy.select() function                                           
housing[' total_bedroom_size']=np.select(case,size)             #syntax=numpy.select(condition,choicelist,default=0)

#np.select function returns array drawn from element in choicelist, depends upon condition .

#output
housing
```
